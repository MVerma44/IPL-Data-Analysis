## Total matches per year

```python
total_matches = df2.groupby(['year'])['date'].count().reset_index().rename(columns={'date':'Matches'})
total_matches
```

```python
plt.figure(figsize=(7,7))
plt.xlabel('Year', fontsize=12)
plt.xticks(fontsize=10)
plt.ylabel('No. of matches', fontsize=12)
plt.xticks(fontsize=10, rotation=90)
plt.title('No. of matches per year', fontsize=15)

plots = sns.barplot(x=total_matches['year'], y=total_matches['Matches'])

for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.0f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0, 8),
                   textcoords='offset points')
        
```

```python
ball_df =  pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
```

```python
ball_df.info()
```

```python
season_data = df2[['id', 'year']].merge(ball_df, left_on='id', right_on='id', how='left').drop('id', axis=1)
season_data.head(5)
```

## Runs per year

```python
runs = season_data.groupby(['year'])['total_runs'].sum().reset_index()
runs
```

```python
plt.xlabel('Year', fontsize=12)
plt.ylabel('Runs Scores', fontsize=12)
plt.title('Total runs scored per year', fontsize=15)
plt.bar(runs['year'], runs['total_runs'], color=['yellowgreen', 'greenyellow', 'palegreen', 'lightgreen', 'darkseagreen'])
```

## Runs scored per team per year

```python
teams = season_data['batting_team'].unique()
teams.sort(axis=0)
teams
```

```python
total_runs = []
for i in range(len(year_list)):
    for j in teams:
        data2 = season_data[season_data['year'] == year_list[i]]
        total = data2[data2['batting_team'] == j]['total_runs'].sum()
        total_runs.append(total)
```

```python
for i in range(len(year_list)):
    for j in teams:    
        datas = season_data[season_data['year'] == year_list[i]]
#         val = [z, teams]
        data_val = pd.DataFrame({'Year':[year_list],'Team':[teams]}) 
```

```python
data_val.explode('Team',ignore_index=True)
a = data_val.explode('Year',ignore_index=True).explode('Team', ignore_index=True)
a['Runs'] = total_runs
a
```

```python
b = a.drop(a[a['Runs'] == 0].index).reset_index().drop('index', axis=1)
b
```

```python
# labels = ['CSK', 'DECCAN', 'DC', 'DD', 'GL', 'PNJ', 'KTK', 'KKR', 'MI', 'PW', 'RR', 'PSG', 'PSGs', 'RCB', 'SRH']

for i in range(len(year_list)):
    x_ = b[b['Year'] == year_list[i]]['Team'].tolist()
    y_ = b[b['Year'] == year_list[i]]['Runs'].tolist()
    
            
    plt.figure(figsize=(6, 6))
    plt.xlabel('Teams', fontsize = 15)
    plt.ylabel('Runs', fontsize = 15)
    plt.xticks(rotation=90)
    plt.title(f'Year: {year_list[i]}', fontsize = 20)
    plots = sns.barplot(x=x_, y=y_)

    for bar in plots.patches:
        plots.annotate(format(bar.get_height(), '.0f'),
                       (bar.get_x() + bar.get_width() / 2,
                        bar.get_height()), ha='center', va='center',
                       size=10, xytext=(0, 8),
                       textcoords='offset points')
```

## Century of players

```python
season_data2 = df2[['id', 'year']].merge(ball_df, left_on='id', right_on='id', how='left')
batsman = season_data2[season_data2['id'] == 335982]['batsman'].unique()
```

```python
cen = season_data2.groupby(['batsman', 'id']).agg({'batsman_runs':'sum'})
cen = cen[cen['batsman_runs'] >= 100]
cen = cen.groupby('batsman').agg({'count'})
cen.columns = cen.columns.droplevel()
cen = cen.sort_values(by='count', ascending=False).reset_index().head()
cen = cen.rename(columns = ({'batsman':'Player', 'count':'Count'}))
cen
```

```python


plt.figure(figsize=(6, 6))
plt.xlabel('Players', fontsize=16)
plt.xticks(rotation=30, fontsize=11)
plt.ylabel('Count', fontsize=16)
plt.title('Most centuries', fontsize=20)


plots = sns.barplot(x=cen['Player'], y = cen['Count'])


for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.0f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=10, xytext=(0, 8),
                   textcoords='offset points')
```
